```python
PATH = '/home/fabio/Documentos/TCC/chest_xray/'
IMG_SIZE = (224,224)
IMG_SHAPE = IMG_SIZE + (3,)
BATCH_SIZE = 16

dados_treinamento = glob.glob(PATH + 'train/**/*.jpeg')
dados_treinamento_normal = glob.glob(PATH + 'train/NORMAL/*.jpeg')
dados_treinamento_pneumonia = glob.glob(PATH + 'train/PNEUMONIA/*.jpeg')
dados_teste = glob.glob(PATH + 'test/**/*.jpeg')
dados_validacao = glob.glob(PATH + 'val/**/*.jpeg')

print("-------------------------------------------------------------------")
print("Total de imagens para treinamento: " + str(len(dados_treinamento)))
print("Total de imagens para teste: " + str(len(dados_teste)))
print("Total de imagens para validação: " + str(len(dados_validacao)))
print("-------------------------------------------------------------------")
```

```python
data_generator = ImageDataGenerator(
    rotation_range=20,
    zoom_range = 0.2,    
    rescale=1/255.)
```

```python
data_generator_test = ImageDataGenerator(
    rescale=1/255.)
```

```python
train_generator = data_generator.flow_from_directory(
    PATH + "train",
    class_mode="categorical",
    shuffle=True,
    batch_size=BATCH_SIZE,
    target_size=IMG_SIZE)

validation_generator = data_generator.flow_from_directory(
    PATH + "val",
    class_mode="categorical",
    shuffle=True,
    batch_size=BATCH_SIZE,
    target_size=IMG_SIZE)

test_generator = data_generator_test.flow_from_directory(
    PATH + "test",
    class_mode="categorical",
    shuffle=True,
    batch_size=BATCH_SIZE,
    target_size=IMG_SIZE)
```

```python
fig=plt.figure(figsize=(15, 10))
columns = 3; rows = 2
for i in range(1, columns*rows +1):
    img, label = train_generator.next()
    fig.add_subplot(rows, columns, i, title=label[i])
    plt.imshow(img[i])
```

Transfer Learning
* Utiliza os pesos da imagenet
* Não inclui a camada densa (totalmente conectada) do modelo

```python
base_model = VGG19(input_shape = IMG_SHAPE, include_top= False, weights = 'imagenet')
base_model.summary()
```

```python
base_model.trainable = True
set_trainable = False
for layer in base_model.layers:
    if layer.name == 'block5_conv1':
        set_trainable = True
    if set_trainable:
        layer.trainable = True
    else:
        layer.trainable = False
```

```python
base_model.summary()
```

* Pooling com media global

```python
model=Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(BatchNormalization())
model.add(Flatten())

model.add(Dense(128, activation="relu", kernel_initializer='he_normal'))
#model.add(Dense(128, activation="relu"))
model.add(Dropout(0.3))
#model.add(layers.Dense(64,activation='relu'))
#model.add(layers.Dense(2,activation='sigmoid'))
model.add(Dense(2,activation='sigmoid'))
base_model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Salvar o modelo com melhor loss durante o treinamento 
checkpointer = ModelCheckpoint(filepath=PATH + "../modelos/model_vgg19.weights.best.hdf5",
                                monitor="val_loss",
                                verbose=1,
                                mode="min",
                                save_best_only=True)

# Interompe o treinamento quando o resultado começar a piorar
early_stop = tf.keras.callbacks.EarlyStopping(monitor="val_loss",
                                min_delta=0.001,
                                patience=5,
                                mode="min",
                                verbose=1)

reduceLROnPlat = ReduceLROnPlateau(monitor='val_loss', 
                                factor=0.8, 
                                patience=10, 
                                verbose=1, 
                                mode='auto', 
                                epsilon=0.0001,
                                cooldown=5,
                                min_lr=0.0001)

history = model.fit(
    train_generator,
    steps_per_epoch = train_generator.samples//BATCH_SIZE, 
    validation_data=validation_generator,
    validation_steps=validation_generator.samples//BATCH_SIZE,
    epochs=50,
    class_weight=class_weight,
    callbacks=[checkpointer,early_stop,reduceLROnPlat])
```

```python
plt.figure(figsize=(8,6))
plt.title('Accuracy scores')
plt.plot(history.history['accuracy'],'go-')
plt.plot(history.history['val_accuracy'],'ro-')
plt.legend(['accuracy', 'val_accuracy'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.title('Loss value')
plt.plot(history.history['loss'],'go-')
plt.plot(history.history['val_loss'],'ro-')
plt.legend(['loss', 'val_loss'])
plt.show()
```

```python
# Carrega o melhor modelo
model = tf.keras.models.load_model(PATH + "../modelos/model_vgg19.weights.best.hdf5")
```

```python
# Carrega as imagens para teste
image_test = np.array([tf.keras.preprocessing.image.img_to_array(tf.keras.preprocessing.image.load_img(image_name, target_size=(IMG_SIZE),color_mode='rgb'))/255 for image_name in dados_teste])
```

```python
# Aplica os testes
y_pred = model.predict(image_test)
y_pred = np.argmax(y_pred, axis=1)
```

```python
# Cria matrix de confusão
conf_matrix = confusion_matrix(test_generator.classes, y_pred)
```

```python
# Imprime a matrix de confusão
plt.figure()
plot_confusion_matrix(conf_matrix,figsize=(12,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.xticks(range(2), ['Normal', 'Pneumonia'], fontsize=16)
plt.yticks(range(2), ['Normal', 'Pneumonia'], fontsize=16)
plt.show()
```

```python
print('Classification Report')
target_names = ['Normal','Pneumonia']
print(classification_report(test_generator.classes, y_pred, target_names=target_names))
```
